```python
# Carregar o CSV
df = pd.read_csv("drive/My Drive/Projeto Futebol/Partidas_Realizadas_Estatisticas/fluXcap.csv", sep=';', header=None)

print(df)

# Renomear as colunas conforme desejado
df.columns = ['valor_mandante', 'estatistica', 'valor_visitante']

#print(df)

# Separar as colunas por tipo de dado: mandante e visitante
df_mandante = df[['estatistica', 'valor_mandante']].set_index('estatistica').T
df_visitante = df[['estatistica', 'valor_visitante']].set_index('estatistica').T

#print(df_mandante)
#print(df_visitante)

# Renomear as colunas para incluir 'mandante' e 'visitante'
df_mandante.columns = [col.lower().replace(' ', '_') + '_mandante' for col in df_mandante.columns]
df_visitante.columns = [col.lower().replace(' ', '_') + '_visitante' for col in df_visitante.columns]

# Concatenar os DataFrames lado a lado, removendo o índice "valor_mandante" e "valor_visitante"
df_final = pd.concat([df_mandante.reset_index(drop=True), df_visitante.reset_index(drop=True)], axis=1)

# Criar a lista de colunas alternadas
colunas_mandante = [col for col in df_final.columns if '_mandante' in col]
colunas_visitante = [col for col in df_final.columns if '_visitante' in col]

# Alternar as colunas mandante e visitante
colunas_alternadas = [val for pair in zip(colunas_mandante, colunas_visitante) for val in pair]

# Reordenar o DataFrame
df_final = df_final[colunas_alternadas]

# Exibir o DataFrame final
df_final
```

```python
import pandas as pd

# Função que determina o resultado da partida
def calcular_resultado(row):
    if row['Placar_Mandante'] > row['Placar_Visitante']:
        return 'VitoriaMandante'  # Vitória Mandante
    elif row['Placar_Mandante'] < row['Placar_Visitante']:
        return 'VitoriaVisitante'  # Vitória Visitante
    else:
        return 'Empate'

# Aplicar a função a cada linha do DataFrame e criar a nova coluna 'resultado'
partidas_realizadas_copia['resultado'] = partidas_realizadas_copia.apply(calcular_resultado, axis=1)

# Visualizar as primeiras linhas para confirmar o resultado
partidas_realizadas_copia.iloc[168]
```

```python
# Mapeando os valores de 'resultado' para números
partidas_realizadas_copia['resultado_numerico'] = partidas_realizadas_copia['resultado'].map({
    'VitoriaMandante': 1,
    'VitoriaVisitante': -1,
    'Empate': 0
})

# Verificar se a conversão foi bem sucedida
print(partidas_realizadas_copia[['resultado', 'resultado_numerico']].head())
```

```python
#partidas_realizadas_copia.dtypes
partidas_realizadas_copia['Placar_Mandante'] = pd.to_numeric(partidas_realizadas_copia['Placar_Mandante'], errors='coerce').astype('Int64')
partidas_realizadas_copia['Placar_Visitante'] = pd.to_numeric(partidas_realizadas_copia['Placar_Visitante'], errors='coerce').astype('Int64')
partidas_realizadas_copia_aux = partidas_realizadas_copia.select_dtypes(include=[float, int])
#partidas_realizadas_copia_aux.corr()
partidas_realizadas_copia_aux.corr()
```
