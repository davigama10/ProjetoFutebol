```python
import numpy as np

# Função para simular o resultado de um jogo com base nas probabilidades
def simular_jogo(prob_vit_mandante, prob_empate, prob_vit_visitante):
    return np.random.choice(
        ['Vitoria_Mandante', 'Empate', 'Vitoria_Visitante'],
        p=[prob_vit_mandante, prob_empate, prob_vit_visitante]
    )
```

```python
def atualizar_pontuacao_classificacao(resultado, mandante, visitante, tabela_classificacao):
    if resultado == 'Vitoria_Mandante':
        tabela_classificacao.loc[tabela_classificacao['Time'] == mandante, 'Pontos'] += 3
        tabela_classificacao.loc[tabela_classificacao['Time'] == mandante, 'Vitorias'] += 1
    elif resultado == 'Empate':
        tabela_classificacao.loc[tabela_classificacao['Time'] == mandante, 'Pontos'] += 1
        tabela_classificacao.loc[tabela_classificacao['Time'] == visitante, 'Pontos'] += 1
        tabela_classificacao.loc[tabela_classificacao['Time'] == mandante, 'Empates'] += 1
        tabela_classificacao.loc[tabela_classificacao['Time'] == visitante, 'Empates'] += 1
    else:
        tabela_classificacao.loc[tabela_classificacao['Time'] == visitante, 'Pontos'] += 3
        tabela_classificacao.loc[tabela_classificacao['Time'] == visitante, 'Vitorias'] += 1

    # Atualizar a coluna Rodada e jogos
    tabela_classificacao.loc[tabela_classificacao['Time'] == mandante, 'Rodada'] += 1
    tabela_classificacao.loc[tabela_classificacao['Time'] == visitante, 'Rodada'] += 1
    tabela_classificacao.loc[tabela_classificacao['Time'] == mandante, 'Jogos'] += 1
    tabela_classificacao.loc[tabela_classificacao['Time'] == visitante, 'Jogos'] += 1
```

```python
tabela_classificacao = pd.read_csv("drive/My Drive/Projeto Futebol/Classificacao_att.csv")

# Função para simular todos os jogos restantes de uma só vez
def simular_todos_os_jogos(probabilidades_jogos_restantes, tabela_classificacao):

    for index, row in probabilidades_jogos_restantes.iterrows():
      resultado_simulado = simular_jogo(row['Prob_Vit_Mandante'], row['Prob_Empate'], row['Prob_Vit_Visitante'])
      atualizar_pontuacao_classificacao(resultado_simulado, row['Mandante'], row['Visitante'], tabela_classificacao)


# Rodar uma simulação completa de todos os jogos
simular_todos_os_jogos(probabilidades_jogos_restantes, tabela_classificacao)

# Filtrar a tabela para mostrar apenas os valores da rodada 38
tabela_final = tabela_classificacao[tabela_classificacao['Rodada'] == 38]

# Ordenar a tabela com base na coluna 'Pontos' e, opcionalmente, 'Vitorias' para critério de desempate
tabela_final = tabela_final.sort_values(by=['Pontos', 'Vitorias'], ascending=[False, False])

# Atualizar a coluna 'Posição' com base na nova ordem
tabela_final['Posicao'] = range(1, len(tabela_final) + 1)

# Exibir a tabela com as posições atualizadas
print(tabela_final)

```

```python
import numpy as np
tabela_classificacao = pd.read_csv("drive/My Drive/Projeto Futebol/Classificacao_att.csv")

def simular_varias_vezes(probabilidades_jogos_restantes, tabela_classificacao, n_simulacoes):
    # Dicionário para armazenar quantas vezes cada time foi campeão
    campeoes = {time: 0 for time in tabela_classificacao['Time']}

    # Executar múltiplas simulações
    for _ in range(n_simulacoes):
        # Copiar a tabela de classificação para não alterar os dados originais
        tabela_simulada = tabela_classificacao.copy()

        # Simular todos os jogos
        simular_todos_os_jogos(probabilidades_jogos_restantes, tabela_simulada)

        # Filtrar a tabela para mostrar apenas os valores da rodada 38
        tabela_final = tabela_simulada[tabela_simulada['Rodada'] == 38]

        # Ordenar a tabela com base nos pontos e, opcionalmente, em critérios de desempate como "Vitorias"
        tabela_final = tabela_final.sort_values(by=['Pontos', 'Vitorias'], ascending=[False, False])

        # Verificar o time campeão (primeiro colocado)
        time_campeao = tabela_final.iloc[0]['Time']
        campeoes[time_campeao] += 1

    # Calcular a probabilidade de cada time ser campeão
    probabilidades_campeao = {time: (campeoes[time] / n_simulacoes) * 100 for time in campeoes}

    return probabilidades_campeao

# Rodar 10.000 simulações (ou mais, dependendo da sua escolha)
probabilidades_finais = simular_varias_vezes(probabilidades_jogos_restantes, tabela_classificacao, n_simulacoes=10000)

# Exibir as probabilidades de cada time ser campeão
print("Probabilidades de cada time ser campeão após simulações:")
for time, probabilidade in probabilidades_finais.items():
    print(f"{time}: {probabilidade:.2f}%")
```
